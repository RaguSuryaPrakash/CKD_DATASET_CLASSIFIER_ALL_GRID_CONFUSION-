```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset=dataset.astype(int)
dataset
```

```python
dataset.columns
```

```python
inde=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
inde
```

```python
dep=dataset[['classification_yes']]
dep
```

```python
dataset['classification_yes'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inde,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
params_grid={"C":[1.0],"class_weight":[None],"multi_class":['auto'],"penalty":['l2']}
grid = GridSearchCV(LogisticRegression(), params_grid, refit = True, verbose = 3,n_jobs=-1) 
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
print("the best parametes:",format(grid.best_params_))
grid_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_pred,average="weighted")
print("the best f1_score:",format(grid.best_params_))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_pred)
```

```python
print("confusion_matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_pred)
```

```python
print("classification_report:\n",clf_report)
```
